```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DBP_df = pd.read_csv('./dataset/DBP.csv')
DGL_df = pd.read_csv('./dataset/DGL.csv')  
DGP_df = pd.read_csv('./dataset/DGP.csv')  
DGZ_df = pd.read_csv('./dataset/DGZ.csv')  
DZZ_df = pd.read_csv('./dataset/DZZ.csv')  
GLD_df = pd.read_csv('./dataset/GLD.csv')  
GLL_df = pd.read_csv('./dataset/GLL.csv')  
IAU_df = pd.read_csv('./dataset/IAU.csv')  
SGOL_df = pd.read_csv('./dataset/SGOL.csv')  
UGL_df = pd.read_csv('./dataset/UGL.csv')

print(len(DBP_df))
print(len(DGL_df))
print(len(DGP_df))
print(len(DGZ_df))
print(len(DZZ_df))
print(len(GLD_df))
print(len(GLL_df))
print(len(IAU_df))
print(len(SGOL_df))
print(len(UGL_df))
```

<h1>Create portfolio DF</h1>

```python
portfolio_df = pd.DataFrame(data = DBP_df['Date'])
portfolio_df['DBP'] = DBP_df['Close']
portfolio_df['DGL'] = DGL_df['Close']
portfolio_df['DGP'] = DGP_df['Close']
portfolio_df['DGZ'] = DGZ_df['Close']
portfolio_df['DZZ'] = DZZ_df['Close']
portfolio_df['GLD'] = GLD_df['Close']
portfolio_df['GLL'] = GLL_df['Close']
portfolio_df['IAU'] = IAU_df['Close']
portfolio_df['SGOL'] = SGOL_df['Close']
portfolio_df['UGL'] = UGL_df['Close']

portfolio_df['Date']= pd.to_datetime(portfolio_df['Date'])

# set index -> data
portfolio_df = portfolio_df.set_index('Date')
portfolio_df
```

<h1>Return Rate of each ETF</h1>

```python
portfolio_return_rate_df = portfolio_df.pct_change()
portfolio_return_rate_df
```

<h1>Find the best weight</h1>

```python
portfolio_weight = []
portfolio_returns = []
portfolio_risk = []

no_etf = len(portfolio_df.columns)
time = 50000

#Find the portfolio risk
#This the year risk rate, so * 252 (one year has 252 exchange day)
variance_matrix = portfolio_return_rate_df.cov()*252    

individual_asset_returns = portfolio_df.resample('Y').last().pct_change().mean()
individual_asset_returns
```

```python
for no in range(time):
    if no % 1000 == 0:
        print(no)
    weight = np.random.random(no_etf)
    weight = weight/np.sum(weight)
    portfolio_weight.append(weight)

    return_rate = np.dot(weight, individual_asset_returns)
    portfolio_returns.append(return_rate)

    #Find the portfolio variance
    var = variance_matrix.mul(weight, axis=0).mul(weight, axis=1).sum().sum()
    #Find the portfolio risk (volatility)
    risk = np.sqrt(var)

    year_risk = risk*np.sqrt(252)
    portfolio_risk.append(year_risk)
```

```python
print(len(portfolio_returns))
print(len(portfolio_risk))
print(len(portfolio_weight))
```

```python
data = {'Return Rate': portfolio_returns, 'Risk':portfolio_risk}

for counter, symbol in enumerate(portfolio_df.columns.tolist()):
    data[symbol + ' weight'] = [weight[counter] for weight in portfolio_weight]

compare_portfolios_df = pd.DataFrame(data)
compare_portfolios_df
```

```python
compare_portfolios_df.plot.scatter(x='Risk', y='Return Rate', marker='o')

plt.xlabel("Risk")
plt.ylabel("Expected Return")
```

<h1>Minimum Risk</h1>

```python
min_risk_portfolio = compare_portfolios_df.iloc[compare_portfolios_df['Risk'].idxmin()]
min_risk_portfolio
```

<h1>Highest Sharpe Ratio</h1>

```python
highest_return_portfolio = compare_portfolios_df.iloc[compare_portfolios_df['Return Rate'].idxmax()]
highest_return_portfolio
```
